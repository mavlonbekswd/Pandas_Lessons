```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 30)
RAW = '../datasets/raw/'
orders = pd.read_csv(RAW+'orders.csv', dtype={'order_id':str,'customer_id':str})
orders['status'] = orders['status'].str.strip().str.lower()
items = pd.read_csv(RAW+'order_items.csv', dtype={'order_id':str,'product_id':str})
items['line_revenue'] = items['quantity']*items['unit_price']*(1-items['discount'])
```

### Warm-Up — NumPy recall (self-check)

```python
arr = np.array([5,150,30,1200,12])
mask = arr > 100
print(int(mask.sum()))
```

### Core lesson tasks

```python
big = (items['quantity'] > 100) & (items['unit_price'] > 50)
print(big.sum())
```

```python
mkt = orders['channel'].isin(['Marketplace','marketplace'])
print(mkt.sum())
```

```python
items['is_suspect'] = (items['quantity'] > 100) | (items['unit_price'].isna())
print(items['is_suspect'].sum())
```

### Mixed review (earlier topics)
