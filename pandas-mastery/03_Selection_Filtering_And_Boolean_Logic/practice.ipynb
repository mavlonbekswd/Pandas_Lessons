## 🕵️ Data Detective Investigation

**Case file #3 — where is revenue leaking?** An auditor wants suspect orders: status in `{cancelled, returned}` AND channel online or marketplace (any casing). What share of all orders is that?

```python
bad_status = orders['status'].isin(['cancelled','returned'])
bad_channel = orders['channel'].isin(['Online','online','Marketplace','marketplace'])
leak_share = (bad_status & bad_channel).mean()
assert 0 < leak_share < 1, 'Share must be a fraction.'
print(f'Suspect orders: {leak_share:.1%}')
```
