```python
def load_aurora(raw):
    ids = ['customer_id','order_id','product_id','ticket_id']
    files = ['customers','products','orders','order_items','returns','marketing_spend','web_traffic','support_tickets','shipments']
    out = {}
    for n in files:
        df = pd.read_csv(raw+n+'.csv')
        for c in ids:
            if c in df.columns: df[c] = df[c].astype(str)
        out[n] = df
    return out

data = load_aurora(RAW)
print({k: v.shape for k, v in data.items()})
```

### Data detective

```python
data = load_aurora(RAW)
biggest = max(data, key=lambda k: len(data[k]))
print(biggest, data['customers']['email'].isna().sum())
```
