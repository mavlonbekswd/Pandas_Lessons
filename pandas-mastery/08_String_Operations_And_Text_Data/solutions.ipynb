```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 30)
RAW = '../datasets/raw/'
tickets = pd.read_csv(RAW+'support_tickets.csv', dtype={'ticket_id':str,'order_id':str})
returns = pd.read_csv(RAW+'returns.csv', dtype={'order_id':str})
```

### Warm-Up — NumPy recall (self-check)

```python
flags = np.array([True, False, True, True])
n = int(flags.sum())
print(n)
```

### Core lesson tasks

```python
tickets['order_ref'] = tickets['message'].str.extract(r'(O\d{6})')
print(tickets['order_ref'].notna().sum())
```

```python
m = tickets['message'].str.lower()
tickets['is_late'] = m.str.contains('late|where is', regex=True, na=False)
tickets['is_damaged'] = m.str.contains('damaged|broken|leak', regex=True, na=False)
tickets['is_refund'] = m.str.contains('refund|return', regex=True, na=False)
print(tickets[['is_late','is_damaged','is_refund']].sum())
```

```python
returns['return_reason'] = returns['return_reason'].str.strip().str.lower()
print(returns['return_reason'].nunique())
```

### Mixed review (earlier topics)

```python
orders = pd.read_csv(RAW+'orders.csv', dtype={'order_id':str})
orders['order_date'] = pd.to_datetime(orders['order_date'], format='mixed', dayfirst=True, errors='coerce')
tk = tickets.merge(orders[['order_id','order_date']], left_on='order_ref', right_on='order_id', how='left')
tk['month'] = tk['order_date'].dt.to_period('M')
print(tk.groupby('month')[['is_late','is_damaged','is_refund']].sum().tail())
```

```python
tickets['ch'] = tickets['channel'].str.lower()
print(tickets.groupby('ch')['is_late'].sum().idxmax())
```
