1. Missing Values Cleaning
Bo'sh qiymatlar bilan ishlash.

Misol:

Name      Age
Ali       25
Mavlon
Sardor    30

Bu yerda Age bo'sh.

Topish: df.isna().sum()

To'ldirish: df.fillna(0) or df.dropna()

2. Duplicate Cleaning

Bir xil rowlar.

Misol:

Ali    25
Ali    25
Mavlon 30

Topish: df.duplicated()

O'chirish: df.drop_duplicates() #extiyot bo'l

5. Invalid Values Cleaning

Format to'g'ri, lekin qiymat noto'g'ri.

Misol:

Age
----
25
30
-10
500

-10 va 500 format jihatdan son.

Lekin mantiqan noto'g'ri.

Topish:

df[(df["Age"] < 0) | (df["Age"] > 120)]

Standartlashtirish:
