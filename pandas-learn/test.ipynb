```python
import pandas as pd

df = pd.read_csv('messy_customers.csv')

print(df.head(30))
```

```python
print("Bo'sh yacheykalar: ",df.isna().sum()) # bo'sh qiymatlarni topish
print("Duplicated:  ", df['Name'].duplicated().sum()) # dublicatedlarni chiqaradi
print("Duplicated:  ", df['Customer_ID'].duplicated().sum()) # dublicatedlarni chiqaradi
```

```python
df['Age'] = pd.to_numeric(
    df['Age'],
    errors='coerce'
)

print( df['Age'].isna().sum())
print(df[df['Age'].isna()])
```

Qaror qabul qilish

1. O'chirish

df.dropna(subset=['Age'])
Variant B

2. Median bilan to'ldirish

df['Age'].fillna(df['Age'].median())
Variant C

3. Agar ma'nosi aniq bo'lsa tuzatish

df['Age'] = df['Age'].replace({
    'twenty': 20,
    'thirty': 30,
    'twenty-six': 26,
    'twenty-eight': 28
})

```python
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
print(df.head(10))
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')

print(df['Join_Date'].head(10))
```

```python
#Avval muammoli rowlarni chiqarardim:
print(df[~df['Name'].str.isalpha().fillna(False)])

#regex bilan tozalsh
df['Name'] = df['Name'].str.replace(
    r'[^a-zA-Z]',
    '',
    regex=True
)
```

```python
df.to_csv('cleaned_customers.csv', index=False)
df.to_excel("cleaned_customers.xlsx",index=False)
```
